# Book Recommendation System
This is the code for a book recommedation system.

```python
# Imports
import os
import pandas as pd

# Read data
books = pd.read_csv("Books.csv")
ratings = pd. read_csv("Ratings.csv")
users = pd.read_csv("Users.csv")
```

```python
# Shape of data
print(books.shape)
print(ratings.shape)
print(users.shape)
```

```python
user_ratings = pd.merge(users, ratings, on="User-ID", how="left")
user_ratings.head(100)
```

```python
# Dropping URLs
books.drop(['Image-URL-S','Image-URL-M','Image-URL-L'],axis=1,inplace=True)
```

```python
# Merging
df = pd.merge(user_ratings,books,on='ISBN',how='left')
df.head(50)
```

```python
# Dropping rows with no reviews
df = df[df['ISBN'].notna()]
df.head(20)
```
